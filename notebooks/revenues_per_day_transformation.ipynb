```python
import pandas as pd
```

```python
revenues_per_day = pd.read_csv('../data/revenues_per_day (1).csv')

print(f'Rows: {revenues_per_day.shape[0]}, columns: {revenues_per_day.shape[1]}')
print(revenues_per_day.dtypes)
```

```python
revenues_per_day.head()
```

#### Convert types

```python
revenues_per_day['date'] = pd.to_datetime(revenues_per_day['date'], errors='coerce')
```

```python
revenues_per_day['theaters'] = pd.to_numeric(revenues_per_day['theaters'], errors='coerce')
revenues_per_day['theaters'] = revenues_per_day['theaters'].round().astype('Int64')
```

```python
print(revenues_per_day.dtypes)
```

#### Standardize categorical text

```python
revenues_per_day['title_raw'] = revenues_per_day['title']
revenues_per_day['distributor_raw'] = revenues_per_day['distributor']
```

```python
revenues_per_day['title'] = revenues_per_day['title'].str.strip().str.lower()
```

```python
revenues_per_day['distributor'] = revenues_per_day['distributor'].astype('string').str.strip()
```

```python
revenues_per_day['distributor'] = revenues_per_day['distributor'].mask(
    revenues_per_day['distributor'].isna() | (revenues_per_day['distributor'] == '-') | (revenues_per_day['distributor'] == ''),
    'Unknown'
)
```

```python
print('Empty titles after strip:', (revenues_per_day['title'].isna() | (revenues_per_day['title'] == '')).sum())
print('Distributor Unknown count:', (revenues_per_day['distributor'] == 'Unknown').sum())
print('Unique titles:', revenues_per_day['title'].nunique(dropna=True))
print('Unique distributors:', revenues_per_day['distributor'].nunique(dropna=True))
```
